# 準備製作dataframe

___

```python

name = []
for url in  c:  
    yt = YouTube(url)
    name.append(yt.title.replace("| 老高與小茉 Mr & Mrs Gao", ""))
    print(len(name),end =" ")
print(name)
```

```python

view = []
for url in  c:  
    yt = YouTube(url)
    view.append(yt.views)
    print(len(view),end =" ")
print(view)
```

```python
length = []
for url in  c:  
    yt = YouTube(url)
    length.append(yt.length)
    print(len(length),end =" ")
print(length)
```

```python
date = []
for url in  c:  
    yt = YouTube(url)
    date.append(yt.publish_date.date())
    print(len(date),end =" ")
print(date)
```

```python
import pandas as pd
import matplotlib.pyplot as plt


yt_data = {"name" :name, 
"view" :view,
"length(sec)" :length,
"date" :date}


df = pd.DataFrame(yt_data)
#df.set_index("date" , inplace=True)
df
```

## 可以直接在pandas結構下進行排序與檢索

```python
df.sort_values(["view"], ascending=False)
```

```python
maxs = df["view"].max()

filt = (df["view"] == maxs)
maxs_name = df.loc[filt]["name"].to_string(index=False)

mins = df["view"].min()
filt1 = (df["view"] == mins) 
mins_name = df.loc[filt1]["name"].to_string(index=False)
    
f"最受歡迎的是{maxs_name},觀看次數{maxs}次;最不受歡迎的是{mins_name},觀看次數{mins}次"
```

```python
df.describe()
#針對點閱數及時常的統計資料
```

```python
df.corr() #觀察時間長度與觀看次數關係，屬於正相關，但相關性不強
```


<font color=#008000>由圖片可見2019年是聲勢最高時期，近幾年則走下坡</font>

```python
data = pd.DataFrame(
    view,
    index=date
    
    #,columns=list("ABCD")
    )
data.cumsum()
data.plot(title="gao's video views",  #圖表標題
                    xlabel='date',  #x軸說明文字
                    ylabel='view(10 million)',
    figsize=(20, 15))
plt.show()
```

## 也可將數據整理至excel方便非工程人員分析


```python
from openpyxl import Workbook

wb = Workbook() #先創建excel檔案
ws = wb.active #打開的第一個工作表

 #將想取得的資訊寫成列表
title = ['標題','觀看數', '時長', '日期']
ws.append(title) #資料標題

for url in  c:
    gao = []
    yt = YouTube(url)
    gao.append(yt.title.replace("| 老高與小茉 Mr & Mrs Gao", ""))
    gao.append(yt.views)
    gao.append(yt.length)
    gao.append(yt.publish_date.date())
    print(gao)
    ws.append(gao) #把列表寫進excel的橫排
        
wb.save('老高頻道數據.xlsx') #寫入後要存檔才會改變    
```
